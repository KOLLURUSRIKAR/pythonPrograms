```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array(['IT','EI','C','ENG','SIN','SEV','EVPR','VDC','PHY'])
ypoints=np.array([93,95,84,90,95,96,92,81,87])
plt.plot(xpoints,ypoints,color='green')
plt.grid(color='g',linestyle='dotted',linewidth=1)
plt.show()
plt.xlabel('subjects')
plt.ylabel('marks')
plt.bar(xpoints,ypoints,color='red',width=0.5)
plt.show()
plt.plot(xpoints,ypoints,'o')
plt.show()
a=['IT','EI','C','ENG','SIN','SEV','EVPR','VDC','PHY']
plt.pie(ypoints,labels=a)
plt.legend()
plt.show()
plt.hist(ypoints)
plt.show()
plt.plot(ypoints,marker='^',color='purple',ms=10)
plt.show()
plt.plot(xpoints,ypoints,"o:r")
plt.show()
plt.barh(xpoints,ypoints,color='c',height=0.5)
plt.show()
```
